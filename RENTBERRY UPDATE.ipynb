```python
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import requests
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
#Activating the chrome browser
driver = webdriver.Chrome()      /html/body/rb-core/div/rb-home-page/rb-promo-search/div/div/div/div/div[2]/rb-geosuggest/div/div/ul/li[2]
```

```python
search_city = driver.find_element(By.XPATH,'/html/body/rb-core/div/rb-home-page/rb-promo-search/div/div/div/div/div[2]/rb-geosuggest/div/div/ul/li[2]')
search_city.click()
```

```python
prop_url = []
start = 0
end = 30
for page in range(start, end):
    url = driver.find_elements(By.XPATH, '//div[@class="property-card-bottom-address rb-icon-before rb-icon--location"]//a')
    for i in url:
        prop_url.append(i.get_attribute('href'))
    next_button = driver.find_elements(By.XPATH, "/html/body/rb-core/div/rb-search-page/rb-search/div/div/rb-search-sidebar/rb-scroll-view/div[1]/rb-filter-bar/rb-sidebar/div/div[2]/rb-search-pagination/div[2]/a[8]")
```

```python
prop_url
```

```python
len(prop_url)
```

```python
Address = [1]
City = [1]
State_code = []
Country = [1]
Rental_price = [1]
Deposit = []
No_of_bed = [1]
No_of_bathroom = [1]
Area = [1]
Pets_allowed = [1]
Washer_dryer = [1]
AC = [1]
Parking = [1]
Fireplace = [1]
Dishwasher = [1]
Hardwood_floors = []
Roofdeck = []
Storage = []
```

```python
# Scrape all the description of property
for url in prop_url:
    driver.get(url)
    time.sleep(1)
    
    try:
        address_element = driver.find_element(By.XPATH, '/html/body/rb-core/div/rb-property/div/div/div/rb-property-main-info/section/div/div/rb-property-data/div/h4/span/span')
        Address.append(address_element.text)
    except NoSuchElementException:
        Address.append('not present')
    
    try:
        city_element = driver.find_element(By.XPATH, '/html/body/rb-core/div/rb-property/div/div/section[2]/div/div[2]/rb-city-statistic/div/div/div[1]/rb-chart-bar/ul/li[2]/span')
        City.append(city_element.text)
    except NoSuchElementException:
        City.append('not present')
    
    try:
        state_code_element = driver.find_element(By.XPATH, '/html/body/div/div/main/div/div/div/div[4]/section[1]/div/div/div[1]/div/div[7]/div/span[2]')
        State_code.append(state_code_element.text)
    except NoSuchElementException:
        State_code.append('not present')
    
    try:
        country_element = driver.find_element(By.XPATH, '/html/body/rb-core/div/rb-property/div/div/section[2]/div/div[2]/rb-city-statistic/div/div/div[1]/rb-chart-bar/ul/li[1]/span')
        Country.append(country_element.text)
    except NoSuchElementException:
        Country.append('not present')
    
    try:
        rental_price_element = driver.find_element(By.XPATH, '/html/body/rb-core/div/rb-property/div/div/div/div[1]/div/div/div/rb-apply-block/div/div/div[1]/div[2]/div')
        Rental_price.append(rental_price_element.text)
    except NoSuchElementException:
        Rental_price.append('not present')
    
    try:
        deposit_element = driver.find_element(By.XPATH, '/html/body/rb-core/div/rb-property/div/div/div/rb-property-main-info/section/div/div/rb-property-facilities/section/div[5]/span[2]')
        Deposit.append(deposit_element.text)
    except NoSuchElementException:
        Deposit.append('not present')
    
    try:
        no_of_bed_element = driver.find_element(By.XPATH, '/html/body/rb-core/div/rb-property/div/div/div/rb-property-main-info/section/div/div/rb-property-facilities/section/div[1]/span[2]')
        No_of_bed.append(no_of_bed_element.text)
    except NoSuchElementException:
        No_of_bed.append('not present')
    
    try:
        no_of_bathroom_element = driver.find_element(By.XPATH, '/html/body/rb-core/div/rb-property/div/div/div/rb-property-main-info/section/div/div/rb-property-facilities/section/div[2]/span[2]')
        No_of_bathroom.append(no_of_bathroom_element.text)
    except NoSuchElementException:
        No_of_bathroom.append('not present')
    
    try:
        area_element = driver.find_element(By.XPATH, '/html/body/rb-core/div/rb-property/div/div/div/rb-property-main-info/section/div/div/rb-property-facilities/section/div[3]/span[2]')
        Area.append(area_element.text)
    except NoSuchElementException:
        Area.append('not present')
    
    try:
        pets_allowed_element = driver.find_element(By.XPATH, '/html/body/rb-core/div/rb-property/div/div/div/rb-property-main-info/section/div/div/rb-property-facilities/section/div[4]/span[3]')
        Pets_allowed.append(pets_allowed_element.text)
    except NoSuchElementException:
        Pets_allowed.append('not present')
    
    try:
        washer_dryer_element = driver.find_element(By.XPATH, '/html/body/rb-core/div/rb-property/div/div/div/div[2]/rb-property-amenities/div/div/div/rb-property-rb-amenities/div/rb-amenities/ul/li[4]/div/span[2]')
        Washer_dryer.append(washer_dryer_element.text)
    except NoSuchElementException:
        Washer_dryer.append('not present')
    
    try:
        ac_element = driver.find_element(By.XPATH, '/html/body/rb-core/div/rb-property/div/div/div/div[2]/rb-property-amenities/div/div/div/rb-property-rb-amenities/div/rb-amenities/ul/li[5]/div/span[2]')
        AC.append(ac_element.text)
    except NoSuchElementException:
        AC.append('not present')
    
    try:
        parking_element = driver.find_element(By.XPATH, '/html/body/rb-core/div/rb-property/div/div/div/div[2]/rb-property-amenities/div/div/div/rb-property-rb-amenities/div/rb-amenities/ul/li[2]/div/span[2]')
        Parking.append(parking_element.text)
    except NoSuchElementException:
        Parking.append('not present')
    
    try:
        fireplace_element = driver.find_element(By.XPATH, '/html/body/rb-core/div/rb-property/div/div/div/div[2]/rb-property-amenities/div/div/div/rb-property-mls-amenities/div/rb-mls-amenities/ul/li[22]/span[2]')
        Fireplace.append(fireplace_element.text)
    except NoSuchElementException:
        Fireplace.append('not present')
    
    try:
        dishwasher_element = driver.find_element(By.XPATH, '/html/body/rb-core/div/rb-property/div/div/div/div[2]/rb-property-amenities/div/div/div/rb-property-rb-amenities/div/rb-amenities/ul/li[5]/div/span[2]')
        Dishwasher.append(dishwasher_element.text)
    except NoSuchElementException:
        Dishwasher.append('not present')
    
    try:
        hardwood_floors_element = driver.find_element(By.XPATH, 'XPATH_FOR_HARDWOOD_FLOORS')
        Hardwood_floors.append(hardwood_floors_element.text)
    except NoSuchElementException:
        Hardwood_floors.append('not present')
    
    try:
        roofdeck_element = driver.find_element(By.XPATH, 'XPATH_FOR_ROOFDECK')
        Roofdeck.append(roofdeck_element.text)
    except NoSuchElementException:
        Roofdeck.append('not present')
    
    try:
        storage_element = driver.find_element(By.XPATH, 'XPATH_FOR_STORAGE')
        Storage.append(storage_element.text)
    except NoSuchElementException:
        Storage.append('not present')
```
